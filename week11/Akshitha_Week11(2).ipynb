```python
import pandas as pd
import time
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
df = pd.read_csv("dfdata.csv")
df.head()
```

```python
df = pd.read_csv("dfdata.csv")
X = df.drop("outcome", axis=1)
y = df["outcome"]

# Define dataset sizes
sizes = [100, 1000, 10000, 100000, 1000000]

print("Size\tAccuracy\tTime (s)")
for size in sizes:
    # If size exceeds dataset, replicate the data
    if size > len(df):
        reps = -(-size // len(df))  # Ceiling division
        df_expanded = pd.concat([df]*reps, ignore_index=True).iloc[:size]
        X = df_expanded.drop("outcome", axis=1)
        y = df_expanded["outcome"]
    else:
        X = df.drop("outcome", axis=1).iloc[:size]
        y = df["outcome"].iloc[:size]

    model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')

    # Time the model fitting and evaluation
    start = time.time()
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    end = time.time()

    print(f"{size}\t{scores.mean():.4f}\t\t{end - start:.2f}")
```
