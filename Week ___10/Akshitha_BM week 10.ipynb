```python
import matplotlib.pyplot as plt
result_scores = {
    'Dummy': (0.4767, 0.5075, 0.5048, 0.5327),
    'KNN Training': (0.8983, 0.895, 0.8867, 0.9045),
    'KNN Prediction': (0.9717, 0.963, 0.9407, 0.9061),
    'KNN Retraining': (0.8983, 0.721, 0.4163, 1.0),
    'KNN Retrained Prediction': (0.9717, 0.965, 0.913, 0.8873)
}

stages = list(result_scores.keys())
accuracy = [result_scores[s][1] for s in stages]
recall = [result_scores[s][3] for s in stages]
precision = [result_scores[s][2] for s in stages]

plt.figure(figsize=(10, 6))
plt.plot(stages, accuracy, marker='o', label='Accuracy')
plt.plot(stages, recall, marker='o', label='Recall')
plt.plot(stages, precision, marker='o', label='Precision')

plt.title('Model Performance Across Stages')
plt.xlabel('Stage')
plt.ylabel('Score')
plt.ylim(0, 1.1)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
